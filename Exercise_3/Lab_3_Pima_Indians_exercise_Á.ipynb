Το αρχείο δεν περιλαμβάνει ονόματα για τα χαρακτηριστικά αλλά μπορείτε να δείτε την περιγραφή τους στο ["Data Set Description"](http://archive.ics.uci.edu/ml/machine-learning-databases/pima-indians-diabetes/pima-indians-diabetes.names).
